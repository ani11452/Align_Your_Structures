```python
import pickle
import numpy as np
import sys
from scipy.spatial.distance import jensenshannon
from scipy.stats import wasserstein_distance
sys.path.append('../../')
# out_path = 'qm9_noH_1000_kabsch_traj_interpolator_finetune_final_539_25_evalout.pkl'
# with open(out_path, 'rb') as f:
#     data = pickle.load(f)
#     print(data.keys())  # smiles
# smiles_1 = list(data.keys())
# print(len(data.keys()))  # 1000

out_path = 'qm9_finetune_interpolation_evalout.pkl'  # place the _evalout.pkl in the working dir
with open(out_path, 'rb') as f:
    data = pickle.load(f)
    print(data.keys())  # smiles
smiles = list(data.keys())
print(len(data.keys()))  # 1000
```

```python
keys = list(data['COCC[C@H]1CC[C@H]1O'].keys())
# for each element in list keys, create a new list to load result
result = [[] for _ in range(len(keys))]
for smile in smiles:
    for key_idx in range(len(keys)):
        key = keys[key_idx]
        if key in data[smile].keys():
            result[key_idx].append(data[smile][key])
```

```python
new_result = []
for result_list in result:
    new_result_list = []
    for data in result_list:
        if not np.isnan(data):
            new_result_list.append(data)
    new_result.append(new_result_list)
result = new_result
```
